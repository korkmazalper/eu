```python
import matplotlib.pyplot as plt  # Für die Grafikdarstellung

def simuliere_sprung(start_hoehe, elastizitaet, zeit_schritte, dt):
    position = start_hoehe
    geschwindigkeit = 0
    gravitation = -9.8
    positionen = []

    for _ in range(zeit_schritte):
        geschwindigkeit += gravitation * dt  # Geschwindigkeit aktualisieren
        position += geschwindigkeit * dt  # Position aktualisieren
        
        # Wenn der Ball den Boden berührt
        if position <= 0:  
            position = 0
            geschwindigkeit = -geschwindigkeit * elastizitaet  # Sprungkraft anpassen
        
        positionen.append(position)
    
    return positionen

# Simulationsparameter
start_hoehe = 10  # Anfangshöhe (Meter)
elastizitaet = 0.8  # Elastizitätskoeffizient (Energieverlust)
zeit_schritte = 200  # Anzahl der Zeitschritte
dt = 0.05  # Zeitintervall (Sekunden)

positionen = simuliere_sprung(start_hoehe, elastizitaet, zeit_schritte, dt)

# Grafik zeichnen
plt.plot([dt * i for i in range(zeit_schritte)], positionen)
plt.xlabel("Zeit (s)")
plt.ylabel("Position (m)")
plt.title("Sprungsimulation eines Balls")
plt.show()
```
